```python
somatic_short = pd.read_csv('somatic_short.maf.gz', sep='\t', compression='gzip')
somatic_short # Read somatic_short data generated from the previous lab session 
```

```python
# Parse the “prot.pos” column to extract the affected protein position (top number) 
# and the protein length (bottom number). 
#Add the parsed output as two new columns to somatic_short. 
prot_pos_parsed = somatic_short['prot.pos'].str.extract('(\d+)/(\d+)') # Matching patterns in a column
prot_pos_parsed
```

```python
prot_pos_parsed.columns = ['prot_pos', 'prot_len'] # Name extracted substring
```

```python
prot_pos_parsed = somatic_short['prot.pos'].str.extract('(?P<prot_pos>\d+)/(?P<prot_len>\d+)')
```

```python
somatic_short = somatic_short.merge(prot_pos_parsed) # Add extracted substrings to a data frame as new columns
```

```python
somatic_short
```

```python
# Parse the “aa” column to extract the reference (top letter)
# and alternative allele (bottom letter). Add the parsed output
# as two new columns to somatic_short
prot_aa_parsed = somatic_short['aa'].str.extract('(\w+)/(\w+)') # Matching patterns in a column
```

```python
prot_aa_parsed
```

```python
prot_aa_parsed.columns = ['top_letter', 'bottom_letter'] # Name extracted substring
```

```python
prot_aa_parsed = somatic_short['aa'].str.extract('(?P<top_letter>\w+)/(?P<bottom_letter>\w+)')
```

```python
somatic_short = somatic_short.join(prot_aa_parsed) # Add extracted substrings to a data frame as new columns
```

```python
somatic_short
```
